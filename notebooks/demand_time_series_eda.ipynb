### Demand stationarity

We then try to test if the time series are stationary or not. This is important for many powerful model to work (ARIMA, SARIMAX).

We first start with an ADF (Augmented Dickey-Fuller) test and KPSS test. These test try to verify two opposite null hypotheses: the first one tests as null hypothesis if the time series is not stationary, the second one extablishes if there is statistical evidence that the time series is stationary.

````python
from statsmodels.tsa.stattools import kpss, adfuller
import warnings
from statsmodels.tools.sm_exceptions import InterpolationWarning

warnings.filterwarnings(
    "ignore",
    category=InterpolationWarning
)

def kpss_test(timeseries, alpha=0.05):
    statistic, p_value, *_ = kpss(
        timeseries,
        regression="c",
        nlags="auto"
    )

    if p_value < alpha:
        print(
            f"KPSS: The series is NOT stationary "
            f"(p-value={p_value:.4f} < alpha={alpha})"
        )
    else:
        print(
            f"KPSS: The series is stationary "
            f"(p-value={p_value:.4f} >= alpha={alpha})"
        )


def adf_test(timeseries, alpha=0.05):
    statistic, p_value, *_ = adfuller(
        timeseries,
        autolag="AIC"
    )

    if p_value < alpha:
        print(
            f"ADF: The series is stationary "
            f"(p-value={p_value:.4f} < alpha={alpha})"
        )
    else:
        print(
            f"ADF: The series is NOT stationary "
            f"(p-value={p_value:.4f} >= alpha={alpha})"
        )


# Example
kpss_test(demand_d1["demand"])
adf_test(demand_d1["demand"])
````

We do not have statistical evidence from either the ADF or KPSS tests to conclude that the time series is stationary.

For the ADF test, the p-value is high, therefore we fail to reject the null hypothesis that the time series is non-stationary.
For the KPSS test, the p-value is smaller than 0.05, therefore we reject the null hypothesis that the time series is stationary.

Both tests suggest that the time series is non-stationary.

````python
for destination in raw_demand['destination_id'].unique():
    print("Destination:", destination)
    kpss_test(raw_demand[raw_demand["destination_id"] == destination]["demand"])
    adf_test(raw_demand[raw_demand["destination_id"] == destination]["demand"])
    print(" ")
````

None of the demand time series is stationary.

### Autocorrelation and partial autocorrelation

````python
import matplotlib.pyplot as plt

from statsmodels.graphics.tsaplots import (
    plot_acf,
    plot_pacf
)

fig, ax = plt.subplots(2, 1, figsize=(12, 8))

plot_acf(
    demand_d1["demand"],
    lags=20,
    ax=ax[0]
)

plot_pacf(
    demand_d1["demand"],
    lags=20,
    ax=ax[1]
)

# Add all x-axis ticks
for axis in ax:
    axis.set_xticks(range(0, 21))
    axis.set_xlabel("Lag")

ax[0].set_ylabel("Autocorrelation")
ax[1].set_ylabel("Partial Autocorrelation")

ax[0].set_title("ACF for demand time series")
ax[1].set_title("PACF for demand time series")

plt.tight_layout()
plt.show()
````

The ACF plot shows significant spikes at lags 7 and 14, suggesting the presence of a weekly seasonal pattern in the demand time series. This indicates that demand values tend to be correlated with observations from the same weekday in previous weeks.

The PACF plot shows significant spikes at lags 1 and 7, suggesting the presence of strong direct temporal dependencies. In particular, the spike at lag 7 further supports the hypothesis of a weekly seasonal structure in the demand time series.

### Decomposition plots

We decompose the demand time series into its main components — trend, seasonality, and residuals. Based on the ACF/PACF analysis, we set the seasonal period to 7.

````python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(
    demand_d1["demand"],
    model="additive",
    period=7
)

fig = result.plot()

fig.set_size_inches(14, 10)

for ax in fig.axes:
    ax.tick_params(axis='x', labelsize=10)
    ax.tick_params(axis='y', labelsize=10)

plt.tight_layout()
plt.show()
````

We check in the following if residuals are normally distributed.

````python
import matplotlib.pyplot as plt
import statsmodels.api as sm

residuals = result.resid.dropna()

fig, ax = plt.subplots(1, 3, figsize=(14, 5))

# Histogram
ax[0].hist(residuals, bins=20)
ax[0].set_title("Residuals Histogram")
ax[0].set_xlabel("Residual")
ax[0].set_ylabel("Frequency")

# QQ Plot
sm.qqplot(
    residuals,
    line="s",
    ax=ax[1]
)

ax[1].set_title("QQ Plot of Residuals")

plt.tight_layout()
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(10, 5))

plot_acf(
    result.resid.dropna(),
    ax=ax
)

plt.show()
````

### Conclusions from the decomposition plots

The decomposition analysis suggests that the demand time series contains:
- a clear increasing trend,
- a strong weekly seasonal component,
- moderate residual noise.

The seasonal component appears highly regular and stable over time, supporting the presence of a strong weekly seasonality in the data.

The residual component appears approximately centered around zero and does not show strong remaining structure, suggesting that trend and seasonality explain a significant portion of the temporal dynamics.

Overall, the decomposition indicates that models capable of explicitly handling:
- trend,
- seasonality,
- temporal dependence

such as ETS/Holt-Winters or SARIMA models may be appropriate forecasting candidates for this dataset.
